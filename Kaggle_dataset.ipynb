```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Students Social Media Addiction.csv')
```

```python
df
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Most_Used_Platform', hue='Academic_Level', data=df)
plt.xlabel('Most_Used_Platform')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Relationship_Status', hue='Academic_Level', data=df)
plt.xlabel('Most_Used_Platform')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
relationship_academic_counts = pd.crosstab(df['Relationship_Status'], df['Academic_Level'])

# Plotting the data
plt.figure(figsize=(10, 7))

# Plot as a stacked bar chart which can sometimes resemble a stacked area plot visually
relationship_academic_counts.plot(kind='bar', stacked=True, ax=plt.gca())

plt.xlabel('Relationship Status')
plt.ylabel('Count')
plt.title('Count of Relationship Status by Academic Level')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
